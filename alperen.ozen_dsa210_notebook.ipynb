```python
# Define start and end dates
start_date = '2020-01-01'
end_date = pd.to_datetime('today').strftime('%Y-%m-%d')

# Download all datasets
df_btc = yf.download('BTC-USD', start=start_date, end=end_date)
df_doge = yf.download('DOGE-USD', start=start_date, end=end_date)
df_shib = yf.download('SHIB-USD', start=start_date, end=end_date)
df_sol = yf.download('SOL-USD', start=start_date, end=end_date)
df_pepe = yf.download('PEPE-USD', start=start_date, end=end_date)

# Reset indices so 'Date' becomes a column
for df in [df_btc, df_doge, df_shib, df_sol, df_pepe]:
    df.reset_index(inplace=True)

# Supply constants
DOGE_SUPPLY = 148_950_000_000        # Dogecoin approximate circulating supply
SHIB_SUPPLY = 589_250_000_000_000     # Shiba Inu approximate circulating supply
PEPE_SUPPLY = 420_690_000_000_000     # Pepe Coin total supply

# Helper function to format columns
def format_columns(df, add_market_cap=False, supply=None):
    df = df[['Date', 'Open', 'High', 'Low', 'Close', 'Volume']]
    if add_market_cap and supply is not None:
        df['Market Cap'] = df['Close'] * supply
    elif add_market_cap:
        df['Market Cap'] = None
    return df

# Apply formatting
df_btc = format_columns(df_btc, add_market_cap=False)  # BTC Market Cap will not be manually calculated now
df_doge = format_columns(df_doge, add_market_cap=True, supply=DOGE_SUPPLY)
df_shib = format_columns(df_shib, add_market_cap=True, supply=SHIB_SUPPLY)
df_sol = format_columns(df_sol, add_market_cap=False)
df_pepe = format_columns(df_pepe, add_market_cap=True, supply=PEPE_SUPPLY)

# Fill missing values with forward-fill
for df in [df_btc, df_doge, df_shib, df_sol, df_pepe]:
    df.fillna(method='ffill', inplace=True)

print(" All datasets are freshly downloaded, Market Caps calculated, and missing values handled.")
```

## Hypotheses Testings

### Creating dataset for analysis

```python
# --- 1. Create Combined Meme Coin Market Cap (DOGE + SHIB only) ---

print(df_doge.shape)
print(df_shib.shape)

# Merge by date (inner join because DOGE and SHIB should overlap)
meme_cap_df = df_doge[['Date', 'Market Cap']].rename(columns={'Market Cap': 'DOGE Market Cap'}) \
    .merge(df_shib[['Date', 'Market Cap']].rename(columns={'Market Cap': 'SHIB Market Cap'}), on='Date', how='inner')

# Create combined meme coin market cap
meme_cap_df['Meme Market Cap'] = meme_cap_df['DOGE Market Cap'] + meme_cap_df['SHIB Market Cap']

print(meme_cap_df.shape)
print("\n")

# --- 2. Prepare BTC & SOL Closing Prices (no Returns yet) ---

print(df_btc.shape)
print(df_sol.shape)

btc_close = df_btc[['Date', 'Close']].rename(columns={'Close': 'BTC Close'})
sol_close = df_sol[['Date', 'Close']].rename(columns={'Close': 'SOL Close'})

# --- 3. Find common date range ---

latest_start = max(
    meme_cap_df['Date'].min(),
    btc_close['Date'].min(),
    sol_close['Date'].min()
)

earliest_end = min(
    meme_cap_df['Date'].max(),
    btc_close['Date'].max(),
    sol_close['Date'].max()
)

print(f"Common date range: {latest_start.date()} to {earliest_end.date()}")

# --- 4. Filter datasets within common date range ---

meme_cap_df = meme_cap_df[(meme_cap_df['Date'] >= latest_start) & (meme_cap_df['Date'] <= earliest_end)]
btc_close = btc_close[(btc_close['Date'] >= latest_start) & (btc_close['Date'] <= earliest_end)]
sol_close = sol_close[(sol_close['Date'] >= latest_start) & (sol_close['Date'] <= earliest_end)]

# --- 5. Merge all together using OUTER JOIN ---

merged_df = meme_cap_df[['Date', 'Meme Market Cap']] \
    .merge(btc_close, on='Date', how='outer') \
    .merge(sol_close, on='Date', how='outer')

# Sort by date
merged_df = merged_df.sort_values(by='Date')

# Forward fill missing Close values (prices)
merged_df[['Meme Market Cap', 'BTC Close', 'SOL Close']] = merged_df[['Meme Market Cap', 'BTC Close', 'SOL Close']].fillna(method='ffill')

# Drop any remaining missing rows (very first rows if any)
merged_df.dropna(inplace=True)

# --- 6. Calculate Returns AFTER merging and filling ---

merged_df['Meme Market Cap Return'] = merged_df['Meme Market Cap'].pct_change()
merged_df['BTC Return'] = merged_df['BTC Close'].pct_change()
merged_df['SOL Return'] = merged_df['SOL Close'].pct_change()

# Drop first row with NaNs (due to pct_change)
merged_df.dropna(inplace=True)

# --- 7. Final Dataset ---

analysis_df = merged_df.copy()

print(" Final analysis_df shape:", analysis_df.shape)
print(analysis_df.head())

# --- 8. Optional: Visualize Correlation Matrix ---

corr = analysis_df.drop(columns=['Date']).corr()

plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix: Meme Market Cap vs BTC & SOL')
plt.show()
```

## Hypothesis 1 (BTC Effect on Meme Cap):

### Null Hypothesis (H) | Bitcoin price movements do not significantly influence Meme Coin Market Cap movements.
### Alternative Hypothesis (H) | Bitcoin price movements positively influence Meme Coin Market Cap movements.

```python
# --- 1. Hypothesis 1: BTC Price Influence on Meme Coin Market Cap ---

# Import library
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Prepare variables
X = analysis_df['BTC Return']  # Independent Variable (Predictor)
y = analysis_df['Meme Market Cap Return']  # Dependent Variable (Target)

# Add constant (bias) term for regression
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the regression summary
print(model.summary())


# Residuals
residuals = model.resid

# Plot residuals
plt.figure(figsize=(10,6))
sns.histplot(residuals, bins=50, kde=True)
plt.title('Distribution of Residuals')
plt.xlabel('Residual')
plt.grid()
plt.show()

# Residuals vs Fitted Plot
fitted_vals = model.predict(X)

plt.figure(figsize=(10,6))
plt.scatter(fitted_vals, residuals, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Fitted Values')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.grid()
plt.show()


```

Based on the regression output:

*   The coefficient for BTC Return is positive and statistically significant (p < 0.001),

*   The model explains 13.1% of the variability in meme coin market cap returns,

*   The p-value of the overall F-test is near zero, indicating the model is a good fit.


 Final Verdict:

*   We reject the null hypothesis (H) and accept the alternative hypothesis (H). There is a statistically significant and positive relationship between Bitcoins price movements and the market capitalization of meme coins.

*   In other words, when Bitcoin performs well, meme coins also tend to gain value, suggesting behavioral or sentiment-driven coupling in the crypto market.










## Hypothesis 2: SOL Price Effect

### Null Hypothesis (H): There is no significant relationship between Solana's price movements and the market capitalization of meme coins.

### Alternative Hypothesis (H): There is a significant positive correlation between Solana's price movements and meme coin market capitalization.


```python
# --- Hypothesis 2: SOL Return  Meme Market Cap Return ---

# Define X (independent variable) and y (dependent variable)
X_sol = analysis_df['SOL Return']  # Predictor: Solana return
y_meme = analysis_df['Meme Market Cap Return']  # Target: Meme cap return

# Add constant to include intercept
X_sol = sm.add_constant(X_sol)

# Fit OLS regression model
model_sol = sm.OLS(y_meme, X_sol).fit()

# Print model summary
print(model_sol.summary())

# --- Residual Analysis ---

# Residuals
residuals_sol = model_sol.resid
fitted_sol = model_sol.predict(X_sol)

# Histogram of residuals
plt.figure(figsize=(10,6))
sns.histplot(residuals_sol, bins=50, kde=True)
plt.title('Distribution of Residuals (SOL Regression)')
plt.xlabel('Residuals')
plt.grid()
plt.show()

# Residuals vs Fitted
plt.figure(figsize=(10,6))
plt.scatter(fitted_sol, residuals_sol, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Fitted Values (SOL Regression)')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.grid()
plt.show()

```

Based on the regression output:

*   The coefficient for SOL Return is positive and statistically significant (p < 0.001),

*   The model explains a modest but real 4.5% of the variation in meme coin market cap return,

*   The null hypothesis is rejected in favor of the alternative.



 Final Verdict:

*   There is a statistically significant and positive relationship between Solanas price movements and the market capitalization of meme coins.

*   However, compared to BTC (which explained 13.1% of meme cap movements), SOL explains less (only 4.5%), indicating that BTC may play a more influential role in driving meme coin market sentiment.












## Hypothesis 3: Combined BTC + SOL Influence

### Null Hypothesis (H): There is no significant relationship between the price movements of Bitcoin (BTC) and Solana (SOL) and the market capitalization of meme coins.

### Alternative Hypothesis (H): Price movements of both Bitcoin and Solana significantly influence the market capitalization of meme coins.

```python
# --- Hypothesis 3: Combined BTC + SOL Return  Meme Market Cap Return ---

# Prepare features and target
X_combined = analysis_df[['BTC Return', 'SOL Return']]  # Multiple independent variables
y_combined = analysis_df['Meme Market Cap Return']       # Dependent variable

# Add constant for intercept
X_combined = sm.add_constant(X_combined)

# Fit multiple linear regression model
model_combined = sm.OLS(y_combined, X_combined).fit()

# Display regression summary
print(model_combined.summary())

# --- Residual Analysis ---

residuals_combined = model_combined.resid
fitted_combined = model_combined.predict(X_combined)

# Histogram of residuals
plt.figure(figsize=(10,6))
sns.histplot(residuals_combined, bins=50, kde=True)
plt.title('Distribution of Residuals (Combined BTC + SOL Model)')
plt.xlabel('Residuals')
plt.grid()
plt.show()

# Residuals vs Fitted plot
plt.figure(figsize=(10,6))
plt.scatter(fitted_combined, residuals_combined, alpha=0.5)
plt.axhline(0, color='red', linestyle='--')
plt.title('Residuals vs Fitted Values (Combined Model)')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.grid()
plt.show()


```

Conclusion:

*   The overall regression model is statistically significant (F-statistic p < 0.001).

*   Bitcoin return has a strong and statistically significant influence on meme coin market cap (p < 0.001).

*   Solana return does not show a statistically significant effect in the presence of BTC (p = 0.378).

*   The combined explanatory power of BTC + SOL returns is about 13.2% of the variability in meme coin returns.



 Final Verdict:

*   We partially reject the null hypothesis.
There is a significant combined effect, but only Bitcoins return is a statistically significant predictor of meme coin market capitalization return.

*   Solanas influence is negligible when Bitcoin is already in the model  this may indicate that Bitcoin drives broader market sentiment, while Solana's influence overlaps or is already captured through BTC movements.
